# Analysis of the database: Information about clients and their respective reservations in a hotel.

```python
import pandas as pd


dtable = pd.read_csv('hotel_bookings.csv')


display(dtable)
```

# Identifying null and duplicate rows in the table.

```python
dtable.info()
print('NULL VALUES\n',dtable.isnull().sum())
print('DUPLICATED VALUES\n',dtable.duplicated().sum())
```

# Cleaning null and duplicated values detected earlier and checking the result.

```python
dtable_cleaned = dtable.drop_duplicates()
print('DUPLICATED VALUES\n',dtable_cleaned.duplicated().sum())

dtable_cleaned = dtable_cleaned.fillna({'country': 'Unknown Country', 'children' : 0, 'agent' : 'Unknown agent', 'company' : 'Unknown company'})
dtable_cleaned['children'] = dtable_cleaned['children'].astype(int)


print('NULL VALUES\n',dtable_cleaned.isnull().sum())
```

# Display of the info and the final result.

```python
dtable_cleaned.info()

display(dtable_cleaned)
```
